# Pandas for Plotting

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.DataFrame(np.random.rand(10,4), columns=['A','B','C','D'])
df
```

# Distribution plot

```python
df.plot.bar();
```

```python
df.plot.bar(stacked=True);
```

## Changing palette

```python
sns.set_palette('magma')
df.plot.bar();
```

## Area plot

```python
df.plot.area();
```

## Changing opacity

```python
df.plot.area(stacked = False)
```

```python
df.plot.area(stacked = False, alpha=0.7)
```

##### Parameter alpha can be changed where 0<alpha<1

## Diff operator

```python
diff_data = df.diff();
diff_data.head(3)
```

## Box-plot

```python
diff_data.plot.box(vert = False,
                  color = { 'medians':'darkblue',
                           'whiskers':'red',
                          'boxes':'purple',
                          'caps':'orange'});
```

## Moving_avg

```python
df1 = pd.DataFrame(np.random.rand(200,1), columns=['value']).reset_index()
df1['value'].plot()
df1['value'].rolling(10).mean().plot();
```

## Distribution plot

```python
df = pd.DataFrame(np.random.rand(50,4), columns=['A','B','C','D'])
```

```python
df.plot.kde();
```

## Scatter plot

```python
df.plot.scatter(x='A',y='B',
                  c='C', #color of data points
                  s=df['C']*100); 
```

## Hexagonal Plot

```python
df.plot.hexbin(x='C',y='D', #hexbin x and y
                 gridsize=18);
```

```python
df.plot.kde();
df.plot.scatter(x='A',y='B',
                  c='C', #color of data points
                  s=df['C']*100); 
df.plot.hexbin(x='C',y='D', #hexbin x and y
                 gridsize=18);
```

## Pie-chart

```python
data = pd.DataFrame(np.random.rand(5,2),
                   index = list("ABCDE"),
                   columns = list("XY"))
data.plot.pie(subplots = True, figsize = (8,4));
```

## Line-plot

```python
df = pd.DataFrame(np.random.rand(100,4), columns=['A','B','C','D'])
df.plot(subplots=True, layout =(2,2), figsize=(20,10));
```

## Coloring

```python
df.plot(subplots=True, figsize=(20,10), color = ['red','blue', 'purple', 'orange']);
```
